### Exhaustive Search Method

```python
from math import *


def f(x):
    f = x**2 + 54 / x
    return f


def exhaustiveSearch(a, b, n):
    counter = 1
    deltaX = (b - a) / n
    x1 = a
    x2 = x1 + deltaX
    x3 = x2 + deltaX

    while x3 <= b:

        print("\nIteration %d)" % (counter))
        print("x1 = %0.5f x2 = %0.5f x3 = %0.5f" % (x1, x2, x3))
        print("f(x1) = %0.5f f(x2) = %0.5f f(x3) = %0.5f" %
              (f(x1), f(x2), f(x3)))

        if f(x1) >= f(x2) and f(x2) <= f(x3):
            return print("\nMinimum lies between %f and %f" % (x1, x3))
        else:
            x1 = x2
            x2 = x3
            x3 = x2 + deltaX
            counter += 1


a = float(input("Enter a: "))
b = float(input("Enter b: "))
n = int(input("Enter no of intermediate points: "))
exhaustiveSearch(a, b, n)
```

### Bounding Phase Method

```python
from math import *


def f(x):
    f = x**2 + 54 / x
    return f


def boundingPhase(X, deltaX):
    k = 0
    previousX = X
    deltaX = abs(deltaX)
    
    print("\nf(X - deltaX) = %0.5f f(X) = %0.5f f(X + deltaX) = %0.5f" % (f(X - deltaX), f(X), f(X + deltaX)))

    if f(X - deltaX) <= f(X) and f(X) <= f(X + deltaX):
        deltaX = -deltaX
        
    print("delta_x = %0.5f" % (deltaX))

    while True:
        newX = X + (2**k)*deltaX
        
        print("\nIteration %d)" % (k + 1))
        print("x = %0.5f f(x) = %0.5f k = %d 2**k*delta = %0.5f" % (X, f(X), k, 2**k*delta))
        print("newX = %0.5f f(newX) = %0.5f" % (newX, f(newX)))

        if f(newX) < f(X):
            k = k+1
            previousX = X
            X = newX
        else:
            return print("\nMinimum lies between %f and %f" % (previousX, newX))


x0 = float(input("Enter initial guess: "))
deltaX = float(input("Enter increment: "))
boundingPhase(x0, deltaX)
```

### Interval Halving Method

```python
from math import *


def f(x):
    f = x**2 + 54 / x
    return f


def intervalHalving(a, b, n):
    counter = 1
    xm = (a + b) / 2

    while counter < n:
        l = b - a
        x1 = a + l / 4
        x2 = b - l / 4

        print("\nIteration %d)" % (counter))
        print("a = %0.5f b = %0.5f l = %0.5f" % (a, b, l))
        print("x1 = %0.5f xm = %0.5f x2 = %0.5f" % (x1, xm, x2))
        print("f(x1) = %0.5f f(xm) = %0.5f f(x2) = %0.5f" % (f(x1), f(xm), f(x2)))

        if f(x1) < f(xm):
            b = xm
            xm = x1
        elif f(x2) < f(xm):
            a = xm
            xm = x2
        else:
            a = x1
            b = x2

        counter += 1

    return print("\nMinimum lies between %f and %f" % (a, b))


a = float(input("Enter a: "))
b = float(input("Enter b: "))
n = int(input("Enter no of iterations: "))
intervalHalving(a, b, n)
```

### Fibonacci Search Method

```python
from math import *


def f(x):
    f = x**2 + 54 / x
    return f


def fibonacci(x):
    if x == 0 or x == 1:
        return 1
    return fibonacci(x - 1) + fibonacci(x - 2)


def fibonacciSearch(a, b, n):
    counter = 1
    l = b - a
    k = 2

    while True:
        lK = (fibonacci(n - k + 1) / fibonacci(n + 1)) * l
        x1 = a + lK
        x2 = b - lK

        print("\nIteration %d)" % (counter))
        print("a = %0.5f b = %0.5f l = %0.5f" % (a, b, l))
        print("k = %0.5f lk = %0.5f x1 = %0.5f x2 = %0.5f" % (k, lK, x1, x2))
        print("f(x1) = %0.5f f(x2) = %0.5f" % (f(x1), f(x2)))

        if f(x1) > f(x2):
            a = x1
        elif f(x1) < f(x2):
            b = x2
        else:
            a = x1
            b = x2

        if k == n:
            break

        k = k + 1
        counter += 1

    return print("Minimum lies between %0.5f and %0.5f" % (a, b))


a = float(input("Enter a: "))
b = float(input("Enter b: "))
n = int(input("Enter no of iterations: "))
fibonacciSearch(a, b, n)
```

### Golden Section Search Method

```python
from math import *


def norm(w):
    x = 5*w
    return x


# function must be normalized
def f(w):
    f = 25*w**2 + 54/(5*w)
    return f


def golden(n):
    counter = 1
    G = 0.618
    a = 0
    b = 1

    while counter <= n:
        w1 = a + G * (b - a)
        w2 = b - G * (b - a)

        print("\nIteration %d)" % (counter))
        print("a = %0.5f b = %0.5f l = %0.5f" % (a, b, b - a))
        print("w1 = %0.5f w2 = %0.5f f(w1) = %0.5f f(w2) = %0.5f" % (w1, w2, f(w1), f(w2)))

        if f(w1) > f(w2):
            if w1 < w2:
                a = w1
            else:
                b = w1
        elif f(w1) < f(w2):
            if w1 < w2:
                b = w2
            else:
                a = w2
        else:
            a = min(w1, w2)
            b = max(w1, w2)

        counter += 1

    return print("\nMinimum lies between %0.5f and %0.5f" % (norm(a), norm(b)))


n = int(input("Enter no of iterations: "))
golden(n)
```

###  Successive Quadratic Estimation Method

```python
from math import *


def f(x):
    f = x**3 - x - 2
    return f


def quad(x0, delta, n):
    counter = 1
    x = [[x0, f(x0)], [x0 + delta, f(x0 + delta)]]

    if x[0][1] >= x[1][1]:
        x.append([x0 + delta*2, f(x0 + delta*2)])
    else:
        x.append([x0 - delta, f(x0 - delta)])

    while counter <= n:
        F_min = x[0][1]
        X_min = x[0][0]
        for i in range(1, len(x)):
            if F_min > x[i][1]:
                F_min = x[i][1]
                X_min = x[i][0]
        
        a1 = (x[1][1] - x[0][1]) / (x[1][0] - x[0][0])
        a2 = (1 / (x[2][0] - x[1][0])) * (((x[2][1] - x[0][1]) / (x[2][0] - x[0][0])) - ((x[1][1] - x[0][1]) / (x[1][0] - x[0][0])))

        print("\nIteration %d)" % (counter))
        print("points:", x)
        print("X_min = %0.5f F_min = %0.5f" % (X_min, F_min))
        print("a1 = %0.5f a2 = %0.5f" % (a1, a2))
        
        if a2 > 0:
            x_dash = ((x[0][0] + x[1][0]) / 2) - (a1 / (2*a2))
            x.append([x_dash, f(x_dash)])
        else:
            return print("a2 =", a2, "< 0")
        
        print("x_dash = %0.5f f(x_dash) = %0.5f" % (x[3][0], x[3][1]))

        for i in range(len(x)):
            for j in range(i + 1, len(x)):
                if x[j][1] < x[i][1]:
                    x[i], x[j] = x[j], x[i]

        del x[3]
        
        for i in range(len(x)):
            for j in range(i + 1, len(x)):
                if x[j][0] < x[i][0]:
                    x[i], x[j] = x[j], x[i]

        counter += 1

    for i in range(len(x)):
            for j in range(i + 1, len(x)):
                if x[j][1] < x[i][1]:
                    x[i], x[j] = x[j], x[i]
    
    print("\nx* = %0.5f \tf(x*) = %0.5f" % (x[0][0], x[0][1]))


x1 = float(input("Enter x1: "))
delta = float(input("Enter delta: "))
n = int(input("Enter no of iterations: "))
quad(x1, delta, n)
```

### Newton Raphson Method 

```python
from math import *


def f(x):
    f = -10 + 2*x + 16/x**2
    return f


def first(x):
    delta = 0.0001
    return (f(x + delta) - f(x - delta)) / (2 * delta)


def second(x):
    delta = 0.0001
    return (f(x + delta) - 2*f(x) + f(x - delta)) / (delta**2)


def newton(x, n):
    counter = 1

    while counter <= n:
        print("\nIteration %d)" % (counter))
        print("x = %0.5f first(x) = %0.5f second(x) = %0.5f" % (x, first(x), second(x)))
        print("f'/f'' = %0.5f" % (first(x) / second(x)))
        x = x - first(x) / second(x)
        
        counter += 1

    print("\nMinimum =", x)


x = float(input("Enter x: "))
n = int(input("Enter no of iterations: "))
newton(x, n)
```

### Bisection Method

```python
from math import *


def f(x):
    f = x**2 + 54/x
    return f


def first(x):
    delta = 0.0001
    return (f(x + delta) - f(x - delta)) / (2 * delta)


def bisection(a, b, n):
    if not first(a) < 0:
        return print("a =", a, " > 0")
    if not first(b) > 0:
        return print("b =", b, " < 0")

    counter = 1

    while counter <= n:
        z = (a + b) / 2
        
        print("\nInteration %d)" % (counter))
        print("a = %0.5f b = %0.5f" % (a, b))
        print("z = %0.5f f'(z) = %0.5f" % (z, first(z)))

        if first(z) < 0:
            a = z
        elif first(z) > 0:
            b = z
        else:
            break

        counter += 1
    
    print("\nmin: (a = %0.5f, b = %0.5f)" % (a, b))


a = float(input("Enter a: "))
b = float(input("Enter b: "))
n = int(input("Enter n: "))
bisection(a, b, n)
```

### Secant Method 

```python
from math import *


def f(x):
    f = x**2 + 54/x
    return f


def first(x):
    delta = 0.0001
    return (f(x + delta) - f(x - delta)) / (2 * delta)


def secant(a, b, n):
    if not first(a) < 0:
        return print("a =", a, " > 0")
    if not first(b) > 0:
        return print("b =", b, " < 0")

    counter = 1

    while counter <= n:
        z = b - first(b) / ((first(b) - first(a)) / (b - a))

        print("\nInteration %d)" % (counter))
        print("a = %0.5f b = %0.5f f'(a) = %0.5f f'(b) = %0.5f" % (a, b, first(a), first(b)))
        print("z = %0.5f f'(z) = %0.5f" % (z, first(z)))

        if first(z) < 0:
            a = z
        elif first(z) > 0:
            b = z
        else:
            break

        counter += 1

    print("\na = %0.5f b = %0.5f minimum = %0.5f" % (a, b, b - first(b) / ((first(b) - first(a)) / (b - a))))


a = float(input("Enter a: "))
b = float(input("Enter b: "))
n = int(input("Enter n: "))
secant(a, b, n)
```

### Cubic Search Method

```python
from math import *


def f(x):
    f = x**2 + 54/x
    return f


def first(x):
    delta = 0.0001
    return (f(x + delta) - f(x - delta)) / (2 * delta)


def cubic(x0, delta, n):
    counter = 1
    
    if first(x0) > 0:
        delta = -delta

    k = 0
    
    while True:
        x_next = x0 + (2**k)*delta
        print("k = %d x = %0.5f x_next = %0.5f" % (k, x0, x_next))

        if first(x_next)*first(x0) < 0:
            x1 = x0
            x2 = x_next
            break
        else:
            k += 1
            x0 = x_next
    
    while counter <= n:
        
        print("\nIteration %d)" % (counter))
        print("x1 = %0.5f x2 = %0.5f" % (x1, x2))
                
        z = 3*(f(x1) - f(x2)) / (x2 - x1) + first(x1) + first(x2)

        w = ((x2 - x1) / abs(x2 - x1)) * (z**2 - first(x1)*first(x2))**0.5
        
        u = (first(x2) + w - z) / (first(x2) - first(x1) + 2*w)
        
        print("z = %0.5f w = %0.5f u = %0.5f" % (z, w, u))
        
        if u == 0:
            x_dash = x2
        elif u > 0 and u <= 1:
            x_dash = x_next - u*(x_next - x0)
        elif u > 1:
            x_dash = x1
            
        print("x_dash = %0.5f f(x_dash) = %0.5f" % (x_dash, f(x_dash)))
            
        while True:
            if f(x_dash) < f(x1):
                break
            else:
                x_dash -= 0.5*(x_dash - x1)
                
        print("f'(x1) = %0.5f f'(x_dash) = %0.5f" % (first(x1), first(x_dash)))
                
        if first(x_dash)*first(x1) < 0:
            x2 = x_dash
        else:
            x1 = x_dash
            
        counter += 1
        
    print("\nMinimum is b/w %0.5f and %0.5f" % (x1, x2))

x0 = float(input("Enter x0: "))
delta = float(input("Enter delta / step-size: "))
n = int(input("Enter n: "))
cubic(x0, delta, n)
```
