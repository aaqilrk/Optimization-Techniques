### Boxâ€™s Evolutionary Optimization Method

```python
from math import *


def f(x, y):
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2


def boxEvolutionary(x1, x2, delta1, delta2, n):

    counter = 1

    while n > 0:

        a1 = x1 - delta1 / 2
        a2 = x2 - delta2 / 2

        b1 = x1 + delta1 / 2
        b2 = x2 - delta2 / 2

        c1 = x1 - delta1 / 2
        c2 = x2 + delta2 / 2

        d1 = x1 + delta1 / 2
        d2 = x2 + delta2 / 2

        print("\nIteration %d)" % (counter))
        print("(delta_1, delta_2) = (%0.5f, %0.5f)" % (delta1, delta2))
        print("(x, y) = (%0.5f, %0.5f) \tf = %0.5f" % (x1, x2, f(x1, x2)))
        
        print("\n(a1, a2) = (%0.5f, %0.5f) \tf = %0.5f" % (a1, a2, f(a1, a2)))
        print("(b1, b2) = (%0.5f, %0.5f) \tf = %0.5f" % (b1, b2, f(b1, b2)))
        print("(c1, c2) = (%0.5f, %0.5f) \tf = %0.5f" % (c1, c2, f(c1, c2)))
        print("(d1, d2) = (%0.5f, %0.5f) \tf = %0.5f" % (d1, d2, f(d1, d2)))

        xm1 = x1
        xm2 = x2

        if f(xm1, xm2) > f(a1, a2):
            xm1 = a1
            xm2 = a2

        if f(xm1, xm2) > f(b1, b2):
            xm1 = b1
            xm2 = b2

        if f(xm1, xm2) > f(c1, c2):
            xm1 = c1
            xm2 = c2

        if f(xm1, xm2) > f(d1, d2):
            xm1 = d1
            xm2 = d2

        n = n - 1
        counter += 1

        if xm1 == x1 and xm2 == x2:
            delta1 = delta1 / 2
            delta2 = delta2 / 2
            continue

        x1 = xm1
        x2 = xm2

    return print("\nMinimum is (%0.5f, %0.5f)" % (x1, x2))


x1 = float(input("Enter x1: "))
x2 = float(input("Enter x2: "))
delta1 = float(input("Enter value of size reduction parameter delta1: "))
delta2 = float(input("Enter value of size reduction parameter delta2: "))
n = int(input("Enter no of iterations: "))
boxEvolutionary(x1, x2, delta1, delta2, n)
```

### Simplex Search Method

```python
from math import *


def f(x, y):
    return x**2 + x*y + 3*y**2 -9*x


def simplexSearch(gamma, beta, n):

    a1 = float(input("Enter a1: "))
    a2 = float(input("Enter a2: "))

    b1 = float(input("Enter b1: "))
    b2 = float(input("Enter b2: "))

    c1 = float(input("Enter c1: "))
    c2 = float(input("Enter c2: "))

    counter = 1

    while counter <= n:

        if(f(a1, a2) > f(b1, b2) and f(a1, a2) > f(c1, c2)):
            xh1 = a1
            xh2 = a2
            if f(b1, b2) > f(c1, c2):
                xg1 = b1
                xg2 = b2
                xl1 = c1
                xl2 = c2
            else:
                xg1 = c1
                xg2 = c2
                xl1 = b1
                xl2 = b2

        elif (f(b1, b2) > f(a1, a2) and f(b1, b2) > f(c1, c2)):
            xh1 = b1
            xh2 = b2
            if f(a1, a2) > f(c1, c2):
                xg1 = a1
                xg2 = a2
                xl1 = c1
                xl2 = c2
            else:
                xg1 = c1
                xg2 = c2
                xl1 = a1
                xl2 = a2

        else:
            xh1 = c1
            xh2 = c2
            if f(a1, a2) > f(b1, b2):
                xg1 = a1
                xg2 = a2
                xl1 = b1
                xl2 = b2
            else:
                xg1 = b1
                xg2 = b2
                xl1 = a1
                xl2 = a2

        print("\nIteration %d)" % (counter))
        print("worst point: xh(%0.5f, %0.5f) f = %0.5f" % (xh1, xh2, f(xh1, xh2)))
        print("next to worst point: xg(%0.5f, %0.5f) f = %0.5f" % (xg1, xg2, f(xg1, xg2)))
        print("best point: xl(%0.5f, %0.5f) f = %0.5f" % (xl1, xl2, f(xl1, xl2)))
        

        xc1 = (xg1 + xl1) / 2
        xc2 = (xg2 + xl2) / 2
        
        print("(xc1, xc2): (%0.5f, %0.5f)" % (xc1, xc2))

        xr1 = 2*xc1 - xh1
        xr2 = 2*xc2 - xh2
        
        print("(xr1, xr2): (%0.5f, %0.5f) f = %0.5f" % (xr1, xr2, f(xr1, xr2)))

        if f(xr1, xr2) < f(xl1, xl2):
            xr1 = (1 + gamma)*xc1 - gamma*xh1
            xr2 = (1 + gamma)*xc2 - gamma*xh2
            print("Expansion")
        elif f(xr1, xr2) >= f(xh1, xh2):
            xr1 = (1 - beta)*xc1 + beta*xh1
            xr2 = (1 - beta)*xc2 + beta*xh2
            print("Contraction")
        elif f(xg1, xg2) < f(xr1, xr2) and f(xr1, xr2) < f(xh1, xh2):
            xr1 = (1 + beta)*xc1 - beta*xh1
            xr2 = (1 + beta)*xc2 - beta*xh2
            print("Contraction")
            
        print("(x_new_1, x_new_2): (%0.5f, %0.5f) f = %0.5f" % (xr1, xr2, f(xr1, xr2)))

        a1 = xr1
        a2 = xr2
        b1 = xg1
        b2 = xg2
        c1 = xl1
        c2 = xl2

        counter += 1

    return print("\nMinimum is (%0.5f, %0.5f)" % (xl1, xl2))


gamma = float(input("Enter gamma: "))
beta = float(input("Enter beta(0, 1): "))
n = int(input("Enter no of iterations: "))
simplexSearch(gamma, beta, n)
```

### Hooke-Jeeves Pattern Search Method

```python
from math import *


def f(point):
    x = point[0]
    y = point[1]
    f = (x**2 + y - 11)**2 + (x + y**2 - 7)**2
    return f


def exploratoryMove(point, delta):

    newPoint = point.copy()

    for i in range(len(point)):

        subList = newPoint.copy()
        subList[i] = subList[i] - delta[i]

        sumList = newPoint.copy()
        sumList[i] = sumList[i] + delta[i]

        if f(subList) <= f(newPoint) and f(subList) <= f(sumList):
            newPoint = subList

        elif f(sumList) <= f(newPoint) and f(sumList) <= f(subList):
            newPoint = sumList

    return newPoint


def patternMove(currentPoint, previousPoint):
    newPoint = []

    for i in range(len(currentPoint)):
        newPoint.append(2 * currentPoint[i] - previousPoint[i])

    return newPoint


def hookeJeeves(point, delta, alpha, n):

    newPoint = []

    while n > 0:
        newPoint = exploratoryMove(point, delta)
        n = n - 1

        if newPoint == point:
            for i in range(len(delta)):
                delta[i] = delta[i] / alpha
        else:
            break

    while n > 0:
        tempPoint = exploratoryMove(patternMove(point, newPoint), delta)
        n = n - 1

        if f(tempPoint) < f(newPoint):
            point = newPoint
            newPoint = tempPoint
        else:
            for i in range(len(delta)):
                delta[i] = delta[i] / alpha

    print("Minimum Point: (%f, %f)" % (newPoint[0], newPoint[1]))


point = []
point.append(float(input("Enter x0: ")))
point.append(float(input("Enter y0: ")))

delta = []
delta.append(float(input("Enter deltaX: ")))
delta.append(float(input("Enter deltaY: ")))

alpha = int(input("Enter alpha: "))

n = int(input("Enter no of iterations: "))

hookeJeeves(point, delta, alpha, n)
```

### Hessian Matrix

```python
import math
from sympy import * 

x, y, x1, x2 = symbols('x y x1 x2')

f = x - y + 2*x**2 + 2*x*y +y**2

print("Expression: {} ".format(f))
print("\nHessian Matrix:")
print("| {} ".format(simplify(Derivative(f, x, x)).doit()), end =" ")
print("{} |".format(simplify(Derivative(f, x, y)).doit()))
print("| {} ".format(simplify(Derivative(f, y, x)).doit()), end =" ")
print("{} |".format(simplify(Derivative(f, y, y)).doit()))
```
